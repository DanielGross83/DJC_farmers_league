```python
url_start = "https://www.worldfootball.net/schedule/"
league_list = ['eng-premier-league-', 'bundesliga-', 'fra-ligue-1-', 'ita-serie-a-', 'ned-eredivisie-', 'esp-primera-division-', 'por-primeira-liga-']
#league_list = ['esp-primera-division-']
url_end = '-spieltag/42/'
start_year = 1963
everything_list = []
for league in league_list:
    year = start_year
    league_dict = {}
    
    if league == 'bundesliga-':
        url_end = '-spieltag/30/'
        
    if league == 'ita-serie-a-':
        url_end = '-spieltag/34/'
        
    if league == 'ned-eredivisie-':
        url_end = '-spieltag/30/'
        
    if league == 'esp-primera-division-':
        url_end = '-spieltag/30/'
        
    if league == 'por-primeira-liga-':
        url_end = '-spieltag/26/'
        
    for n in range(60):
        
        end_year = year + 1
        url_year = str(year) + '-' + str(end_year)
        weird_url = False
        
        if league == 'eng-premier-league-':
            if end_year == 1989:
                url_end = '-spieltag/38/'
            elif end_year == 1993:
                url_end = '-spieltag/42/'
            elif end_year == 1996:
                url_end = '-spieltag/38/'
                
        if league == 'bundesliga-':
            if end_year == 1966:
                url_end = '-spieltag/34/'
            elif end_year == 1992:
                url_end = '-spieltag/38/'
            elif end_year == 1993:
                url_end = '-spieltag/34/'
                
        if league == 'fra-ligue-1-':
            if end_year == 1966:
                url_end = '-spieltag/38/'
            elif end_year == 1969:
                url_end = '-spieltag/34/'
            elif end_year == 1971:
                url_end = '-spieltag/38/'
            elif end_year == 1998:
                url_end = '-spieltag/34/'
            elif end_year == 2003:
                url_end = '-spieltag/38/'
                
        if league == 'ita-serie-a-':
            if end_year == 1968:
                url_end = '-spieltag/30/'
            elif end_year == 1989:
                url_end = '-spieltag/34/'
            elif end_year == 2005:
                url_end = '-spieltag/38/'
                
        if league == 'ned-eredivisie-':
            if end_year == 1967:
                url_end = '-spieltag/34/'
                
        if league == 'esp-primera-division-':
            if end_year == 1972:
                url_end = '-spieltag/34/'
            elif end_year == 1987:
                weird_url = True
            elif end_year == 1988:
                url_end = '-spieltag/38/'
            elif end_year == 1996:
                url_end = '-spieltag/42/'
            elif end_year == 1998:
                url_end = '-spieltag/38/'
            elif end_year == 2017:
                weird_url = True
                
        if league == 'por-primeira-liga-':
            if end_year == 1972:
                url_end = '-spieltag/30/'
            elif end_year == 1988:
                url_end = '-spieltag/38/'
            elif end_year == 1990:
                url_end = '-spieltag/34/'
            elif end_year == 1991:
                url_end = '-spieltag/38/'
            elif end_year == 1992:
                url_end = '-spieltag/34/'
            elif end_year == 2003:
                league = 'por-superliga-'
        if league == 'por-superliga-':
            if end_year == 2007:
                url_end = '-spieltag/30/'
            if end_year == 2011:
                league = 'por-liga-sagres-'
        if league == 'por-liga-sagres-':
            if end_year >= 2012:
                league = 'por-liga-zon-sagres-'
        if league == 'por-liga-zon-sagres-':
            if end_year >= 2014:
                league = 'por-primeira-liga-'
        if league == 'por-primeira-liga-':
            if end_year == 2015:
                url_end = '-spieltag/34/'
        
        url = url_start + league + url_year + url_end
        
        if weird_url:
            if league == 'esp-primera-division-':
                if end_year == 1987:
                    url = 'https://www.worldfootball.net/schedule/esp-primera-division-1986-1987-playoff-1-6-spieltag/44/'
                if end_year == 2017:
                    url = 'https://www.worldfootball.net/schedule/esp-primera-division-2016-2017-spieltag_2/38/'
        
        page = requests.get(url)
        soup = BeautifulSoup(page.content, "html.parser")
        table = soup.find_all('table', class_='standard_tabelle')
        interest_table = table[1]
        header = []
        rows = []
        for i, row in enumerate(interest_table.find_all('tr')):
            if i == 0:
                header = [el.text.strip() for el in row.find_all('th')]
            else:
                rows.append([el.text.strip() for el in row.find_all('td')])
                
        year = year + 1
#        print(url_year)
#        print(rows[0])
#        print(url)
        league_dict[url_year] = {}
        league_dict[url_year]['teams_ordered'] = [i[2] for i in rows]
        league_dict[url_year]['champion'] = rows[0][2] 
        league_dict[url_year]['champ_points'] = int(rows[0][9].split(':')[0])
        league_dict[url_year]['champ_margin'] = int(rows[0][9].split(':')[0]) - int(rows[1][9].split(':')[0])
    everything_list += [league_dict]
    print('one league done, kill me now')
        
    #print(league_dict)
    
print(everything_list)
```

```python
import pandas as pd
import pathlib 
for j in range(len(everything_list)):
    league_data = everything_list[j]
    league_name = league_list[j]
    league_name = league_name.replace('-', '')
    csv_name = league_name + '.csv'
    for i in league_data:
        season = league_data[i]
        season["year"] = i
        contents = pd.DataFrame([season])
        p = pathlib.Path(csv_name)
        if p.exists():
            mode = 'a'
            header = False
        else:
            mode = 'w'
            header = True
        contents.to_csv(p, mode=mode, header=header, index=False)
```
